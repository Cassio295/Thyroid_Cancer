```python
import asyncio
from bitrix24 import Bitrix24 
bx24 = Bitrix24('https://b24-da84sm.bitrix24.com.br/rest/163/j1cn4wbe3xv3tagq')
response = await asyncio.wait_for (bx24.callMethod('tasks.task.list', 
                                 filter = {'>CREATED_DATE': '2024-06-01',
                                           'RESPONSIBLE_ID': 43 },
                                 select = ['ID', 'TITTLE', 'RESPONSIBLE_ID', 'CREATED_DATE' ]), timeout=100)
```

```python
import pandas as pd 
tarefas_nilo = pd.DataFrame(response['tasks'])
tarefas_nilo['id'] = tarefas_nilo['id'].astype('int32')
tarefas_nilo
```

```python
tarefas = [tarefa for tarefa in tarefas_nilo['id']]
resultado_hist = []

for t in tarefas:
    historico =await bx24.callMethod('tasks.task.history.list', taskId = t)
    resultado_hist.append(historico)
```

```python

import datetime 
import pytz
final = pd.DataFrame([i for n in resultado_hist for i in n['list']])
prazo_alterado = final.loc[final['field'] == 'DEADLINE']
prazo_alterado['antigo'] = [n['from'] for n in prazo_alterado['value']]
prazo_alterado['novo'] = [n['to'] for n in prazo_alterado['value']]
prazo_alterado = prazo_alterado.drop('value', axis = 1)
prazo_alterado['antigo'] = pd.to_numeric(prazo_alterado['antigo'], errors='coerce')
prazo_alterado['novo'] = pd.to_numeric(prazo_alterado['novo'], errors='coerce')
final
```

```python

prazo_alterado.fillna(0)
prazo_alterado['Data Formatada'] = pd.Series(dtype=object)
fuso = pytz.timezone('America/Sao_Paulo')
for index, row in prazo_alterado.iterrows():
    time = row['novo']
    datas_br = []
    if time is not None:
        data_hora = datetime.datetime.fromtimestamp(time,fuso)
        data_format = data_hora.strftime('%d/%m/%Y/')
        datas_br.append(data_format)
    prazo_alterado.loc[index, 'Data Formatada'] = datas_br
prazo_alterado.info()
```
