```python
# Carregando o arquivo de texto que vou usar
import nltk as n
from nltk.corpus import stopwords
arquivo = open('Memórias-Postumas-de-Brás-Cubas-PDF.txt', 'r', encoding = 'utf-8').read()
# Eliminando os caracteres desnecessarios
tokenizer = n.RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(arquivo)
stop = stopwords.words('portuguese')
palavras_sem_stopwords = [token.lower() for token in tokens if token not in stop]



```

```python
# extraindo os bigramas, Trigramas e n-gramas
from nltk import trigrams, bigrams, ngrams
bigrams = bigrams(palavras_sem_stopwords)
trigrams = trigrams(palavras_sem_stopwords)
ngrams = ngrams(palavras_sem_stopwords,4)
list(ngrams)
```

```python
# Criando os etiquetadores
from nltk.corpus import mac_morpho
from nltk.tag import DefaultTagger, UnigramTagger 
etiq_padrao = DefaultTagger('N')
setencas_treinadas = mac_morpho.tagged_sents()
etiq = UnigramTagger(setencas_treinadas, backoff= etiq_padrao)
tags = etiq.tag(tokens)
```

```python
from nltk.chunk import RegexpParser
pattern = 'NP: {<NPROP><NPROP>|<N><N>}'
analise_gramatical = RegexpParser(pattern)
arvore = analise_gramatical.parse(tags)
print(arvore)
```
